```python
import pandas as pd
import numpy as np
```

Download .tsv file from https://www.imdb.com/interfaces/ \
we download title.basics.tsv and title.ratings.tsv \
and save the data in data folder



```python
title_ratings = pd.read_csv("data/title.ratings.tsv", sep='\t')
# title_ratings = pd.read_csv("data/title.ratings.csv")
title_ratings.head()
```

```python
title_basics = pd.read_csv("data/title.basics.tsv", sep='\t')
# title_basics = pd.read_csv("title.basics.csv")
```

```python
title_basics = title_basics[['titleType','tconst','primaryTitle', 'originalTitle', 'startYear']]
# title_basics = title_basics[title_basics.titleType=='movie']
title_basics = title_basics[title_basics.startYear.apply(lambda x: str(x).isnumeric())]
title_basics.head()
```

```python
print(title_ratings.shape)
print(title_basics.shape)
```

```python
# title_ratings.to_csv('title.ratings.csv', index=False)
# title_basics.to_csv('title.basics.csv', index=False)
```

```python
ratings_and_titles = pd.merge(title_ratings.set_index('tconst'), title_basics.set_index('tconst'), left_index=True, right_index=True, how='inner')
ratings_and_titles = ratings_and_titles.drop_duplicates()
ratings_and_titles
```

we download netflix_titles.csv from https://www.kaggle.com/datasets/shivamb/netflix-shows

```python
netflix_titles = pd.read_csv("data/netflix_titles.csv")
print(netflix_titles.shape)
netflix_titles = netflix_titles.dropna(subset=['release_year'])
netflix_titles.release_year = netflix_titles.release_year.astype(np.int64)
print(netflix_titles.shape)
netflix_titles.head()
```

```python
ratings_and_titles = ratings_and_titles[ratings_and_titles.startYear.apply(lambda x: str(x).isnumeric())]
ratings_and_titles.startYear = ratings_and_titles.startYear.astype(np.int64)
ratings_and_titles.head()
```

```python
netflix_titles['title_low'] = netflix_titles['title'].str.lower()
ratings_and_titles['originalTitle_low'] = ratings_and_titles['originalTitle'].str.lower()
ratings_and_titles['primaryTitle_low'] = ratings_and_titles['primaryTitle'].str.lower()
```

```python
ratings_and_titles.head()
```

One movie can have multiple year of reviews. So, we have to average it.

```python
ratings_and_titles[ratings_and_titles['primaryTitle']=='Breaking Bad']
```

```python
ratings_and_titles['all_averageRating'] = ratings_and_titles['averageRating'] * ratings_and_titles['numVotes']
ratings_and_titles.head()
```

```python
grouped_ratings_and_titles = ratings_and_titles.groupby(by=['primaryTitle_low'])[['numVotes', 'all_averageRating']].sum().reset_index()
grouped_ratings_and_titles['averageRating'] = grouped_ratings_and_titles[['all_averageRating']].div(grouped_ratings_and_titles.numVotes, axis=0)
grouped_ratings_and_titles['averageRating'] = grouped_ratings_and_titles['averageRating'].round(1)

grouped_ratings_and_titles
```

```python
final = pd.merge(netflix_titles, grouped_ratings_and_titles, left_on=['title_low'], right_on=['primaryTitle_low'], how='left')
final = final.drop(['title_low', 'primaryTitle_low', 'all_averageRating'], axis=1)
final = final.rename(columns={'averageRating':'IMDB_rating', 'numVotes':'IMDB_numvotes'})
print(final.shape)
final.head()
```

This is our final dataset

```python
final.to_csv("data/netflix_titles_with_IMDB.csv", index=False)
```
